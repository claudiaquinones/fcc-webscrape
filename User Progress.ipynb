```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
labels = ['Complete', 'Incomplete']
titles = ['Exercises', 'Projects Created', 'Certifications']
colors = ['lightgreen', 'lightgray']
```

```python
user_overview = pd.read_csv('userOverview.csv')
user_overview.head()
```

```python
index = 0

for index in range(len(user_overview)):
    sizes = [user_overview.Complete[index], user_overview.Incomplete[index]]

    plt.pie(sizes, labels=labels, colors=colors,
            autopct='%1.1f%%', shadow=True, startangle=140)

    plt.axis('equal')
    plt.title(titles[index])
    plt.show()
    index += 1
```

```python
user_curriculum = pd.read_csv("userCurriculum.csv")
user_curriculum.head(10)
```

```python
index = 0

for index in range(len(user_curriculum)):
    sizes = [user_curriculum.Completed[index], user_curriculum.Challenges[index] - user_curriculum.Completed[index]]

    plt.pie(sizes, labels=labels, colors=colors,
            autopct='%1.1f%%', shadow=True, startangle=140)

    plt.axis('equal')
    plt.legend()
    plt.title(user_curriculum.Topic[index] + ": " + str(user_curriculum.Challenges[index]) + " Total Challenges")
    plt.show()
    index += 1
```
